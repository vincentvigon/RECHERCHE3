```python
import tensorflow as tf
import numpy as np
```

```python
@tf.function(input_signature=(tf.TensorSpec(shape=[None,2], dtype=tf.float32),))
def test_signature(W):
    print("trace avec W.shape:",W.shape)
    return W*2
```

```python
test_signature(tf.ones([2,2]))
```

```python
test_signature(tf.zeros([2,2]))
```

```python
test_signature(tf.ones([3,2]))
```

```python
test_signature(tf.ones([3,3]))
```
